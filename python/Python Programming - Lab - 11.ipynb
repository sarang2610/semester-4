```python
x = range(1,11,1)
cxMarks= [8,9,6,3,5,7,4,1,2,6]
cyMarks= [5,8,9,6,3,2,4,8,8,9]


# write a code to generate below graph
plt.plot(x,cxMarks,marker="o",color='r',ls='--')
plt.plot(x,cyMarks,marker="^",color='b',ls='--')
plt.annotate(xy=[8,1],text='lowest spi')
plt.legend(['cx','cy'],loc=3)
plt.xlabel="roll no"
plt.ylabel="spi"
plt.show()
plt.savefig('mychart.jpg')
```

### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40]
y=['food','travel','acomodotation','shoping']
c=['r','g','y','m']
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=ex,shadow=True,autopct="%0.1f%%",textprops={"fontsize":15,"color":"b"},labeldistance=1.3)
plt.title("Expencis")
```

### 03) WAP to demonstrate the use of Bar chart.

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["python","c++","c","java"]
y=[85,70,60,82]
c=['r','g','y','m']
def addLables(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
        
plt.bar(x,y,color=c,edgecolor="b",alpha=0.7)
addLables(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
c=['b','r','m','y','r']
s=[10,13,15,12,17]
plt.scatter(x,y,color=c,marker='*',s=s)
plt.xlabel("x labe")
plt.ylabel("y label")
plt.show()
```
